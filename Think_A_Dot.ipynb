```python
class Dot:
    
    def __init__(self, exit_yellow = None, exit_blue = None):
        self.colors = ["Y", "B"]
        self.color = 0
        self.exit = [ exit_yellow, exit_blue]
        
    def set_color(self,color):
        self.color = self.colors.index(color)
        
    def get_color(self):
        return self.colors[self.color]
    
    def set_exit(self,yellow_exit, blue_exit):
        self.exit = [ yellow_exit, blue_exit ]
        
    def drop(self):
        self.color = 1 - self.color
        if self.exit[self.color] != None:
            self.exit[self.color].drop()

class ThinkADot:
    
    def __init__(self):
        self.dots = [[None,None,None],[None,None],[None,None,None]]
        for i in range(len(self.dots)):
            for j in range(len(self.dots[i])):
                self.dots[i][j] = Dot()
                
        self.dots[1][0].set_exit(self.dots[0][0],self.dots[0][1])
        self.dots[1][1].set_exit(self.dots[0][1],self.dots[0][2])
        
        self.dots[2][0].set_exit(self.dots[1][0],self.dots[0][0])
        self.dots[2][1].set_exit(self.dots[1][0],self.dots[1][1])
        self.dots[2][2].set_exit(self.dots[0][2],self.dots[1][1])
    
    def set_colors(self, blue, yellow):
        for p in blue:
            self.dots[p[0]][p[1]].set_color('B')
        for p in yellow:
            self.dots[p[0]][p[1]].set_color('Y')
       
    def left_tilt(self):
        set_blue = [(2,0),(2,2),(0,0),(0,2)]
        set_yellow = [(2,1),(1,0),(1,1),(0,1)]
        self.set_colors(set_blue, set_yellow)

    def right_tilt(self):
        set_yellow = [(2,0),(2,2),(0,0),(0,2)]
        set_blue = [(2,1),(1,0),(1,1),(0,1)]
        self.set_colors(set_blue, set_yellow)

    def drop(self,hole):
        assert hole>=0 and hole<3, 'hole out of range'
        self.dots[2][hole].drop()
        
    def drop_sequence(self,hole_sequence, draw_all=False):
        for d in hole_sequence:
            self.drop(d)
            if draw_all:
                self.draw()
                print()
        
    def draw(self):
        print(f'{self.dots[2][0].get_color()} {self.dots[2][1].get_color()} {self.dots[2][2].get_color()}\n'
              +f' {self.dots[1][0].get_color()} {self.dots[1][1].get_color()}\n'
              +f'{self.dots[0][0].get_color()} {self.dots[0][1].get_color()} {self.dots[0][2].get_color()}')
```

```python

def from_left_to_all_yellow(tad):
    sequence = [0,1,1,2,2,2]
    
    tad.left_tilt()
    tad.draw()
    print()
    tad.drop_sequence(sequence, draw_all=True)

def from_right_to_all_blue(tad):
    sequence = [2,1,1,0,0,0]
    
    tad.right_tilt()
    tad.draw()
    print()
    tad.drop_sequence(sequence, draw_all=True)

def from_right_to_left(tad):
    sequence = [0,1,1,1,1,1,2]
    
    tad.right_tilt()
    tad.draw()
    print()
    tad.drop_sequence(sequence, draw_all=True)
    
def from_left_to_right(tad):
    sequence = [0,1,1,1,1,1,2]
    
    tad.left_tilt()
    tad.draw()
    print()
    tad.drop_sequence(sequence, draw_all=True)

tad = ThinkADot()
from_left_to_all_yellow(tad)
print()
from_right_to_all_blue(tad)
print()
from_right_to_left(tad)
print()
from_left_to_right(tad)
```
